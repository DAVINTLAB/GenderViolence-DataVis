Script para as seguintes padronizações:
1) Nas colunas "País", "Nacionalidade da vítima", padroniza o nomes dos paises. Alguns aparecem mais de um vez em razão de haver uma abreviação de seu nome. Exemplo: muda de "AR | ARGENTINA" para "ARGENTINA".
2) Nas colunas "Município", "Município da vítima", padroniza o nome dos municipíos. Alguns aparecem mais de uma vez em razão de haver um código antes do nome ou haver ocorrências com e sem acento. Exemplo: muda de "110001 | ALTA FLORESTA" para "ALTA FLORESTA".
3) Padronização das profissões segundo dicionário do ENEM.


```python
import pandas as pd
import unidecode
```

```python
# Lê o arquivo
nomeDoArquivo = "2020-2 a 2023"
data = pd.read_excel(nomeDoArquivo + ".xlsx")
# Imprime o nome das colunas
colunas = data.columns.sort_values()
for coluna in colunas:
    print(coluna)
```

Padronização dos nomes de países e munícipio

```python
def padronizar_pais(pais):
    if isinstance(pais, str) and '|' in pais:
        pais_limpo = pais.split('|')[-1].strip()
        pais_limpo = unidecode.unidecode(pais_limpo)  # Remove acentos e caracteres especiais
        return pais_limpo.upper()  # Converte para caixa alta
    return pais

def padronizar_municipio(municipio):
    if isinstance(municipio, str):
        municipio_limpo = municipio.split('|')[-1].strip()
        municipio_limpo = unidecode.unidecode(municipio_limpo)  # Remove acentos e caracteres especiais
        return municipio_limpo.upper()  # Converte para caixa alta
    return municipio

# Padronizando os nomes dos países
data["País"] = data["País"].apply(padronizar_pais)
data["Nacionalidade da vítima"] = data["Nacionalidade da vítima"].apply(padronizar_pais)

# Padronizando os nomes dos municipios
data["Município"] = data["Município"].apply(padronizar_municipio)
data["Município da vítima"] = data["Município da vítima"].apply(padronizar_municipio)
    
```

```python
# Faz uma cópia para posterior comparação.
dataCopia = data.copy()

# Definição da função para verificar se alguma profissão do grupo está contida na profissão do suspeito ou vitima. 
# A comparação é feita verificando se a string da profissão de cada grupo está contida dentro da string da profissão da tabela.
def classifica(profissao):
    if not isinstance(profissao, str):
        return profissao
    
    # Remoção de acentos e conversão para minusculas.
    profissao = unidecode.unidecode(profissao.lower())
    
    if profissao.lower == "estudante":
        return "ESTUDANTE" 
    
    for radical in profissoes_grupo5:
        if radical in profissao:
            return "GRUPO 5"
        
    for radical in profissoes_grupo4:
        if radical in profissao:
            return "GRUPO 4"
    
    for radical in profissoes_grupo2:
        if radical in profissao:
            return "GRUPO 2"
        
    for radical in profissoes_grupo3:
        if radical in profissao:
            return "GRUPO 3"
        
    for radical in profissoes_grupo1:
        if radical in profissao:
            return "GRUPO 1"
        
    return "OUTROS"

# Aplica a função de classificação nas colunas
data["Profissão do suspeito"] = data["Profissão do suspeito"].apply(classifica)
data["Profissão da vítima"] = data["Profissão da vítima"].apply(classifica)
```
