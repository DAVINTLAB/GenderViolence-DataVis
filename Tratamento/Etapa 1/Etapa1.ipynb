Código para:

    1) Padronização dos nomes das colunas (baseado no arquivo "Comparação dos campos de cada tabela.xlsx");

    2) Filtragem por violência doméstica e remoção de duplicatas (apenas nos dados de 2020 a 2023 é necessario);

    3) Padronização para todos os valores de todas as tabelas estarem em caixa alta e sem acento;
    
    4) Conversão de .csv para .xlsx;

OBS: em razão dos arquivos com os dados serem muito pesados, não é possível disponibilizá-los no GitHub, mas basta fazer o download dos arquivos no link abaixo e inseri-los na pasta "Etapa 1".

https://www.gov.br/mdh/pt-br/acesso-a-informacao/dados-abertos/ligue180

São usados nesse processamento os arquivos de 2014 até o segundo semestre de 2023 (inclusive ele), com excessão do "Balanço 2019".

```python
import pandas as pd
from unidecode import unidecode
```

```python
def RemoveAcentosCaixaAlta(celula):
    if isinstance(celula, str):
        return unidecode(celula).upper()
    return celula

def Processa(nomeCSV, nomeXLSX, tabelaComNomesDasColunas, delimitador):
    # Le o arquivo CSV
    tabela = pd.read_csv(nomeCSV, delimiter=delimitador, dtype=str, on_bad_lines='skip')

    # Padroniza o nome das colunas e filtra os dados
    if nomeCSV == 'ligue180-2014.csv': 
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[0].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas

    if nomeCSV == 'ligue180-2015.csv': 
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[1].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas

    if nomeCSV == 'ligue180-2016.csv': 
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[2].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas

    if nomeCSV == 'ligue180-2017.csv': 
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[3].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas

    if nomeCSV == 'ligue180-nov-2018.csv': 
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[4].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas

    if nomeCSV == 'ligue180-2018.csv': 
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[5].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas

    if nomeCSV == 'ligue180-2019.csv': 
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[6].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas

    if nomeCSV == 'primeiro-semestre-2020.csv': 
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[7].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas

        tabela = tabela[tabela['Grupo vulnerável'].isin(["1-Violência Contra a Mulher", "2-Violência Doméstica e Familiar Contra a Mulher"])]

    if nomeCSV == 'segundo-semestre-2020.csv':
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[8].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas
        tabela = tabela.drop_duplicates(subset='Hash', keep='first')

        tabela = tabela[tabela['Grupo vulnerável'].isin(["01. OUTRAS VIOLÊNCIAS CONTRA A MULHER", "02. VIOLÊNCIA DOMÉSTICA E FAMILIAR CONTRA A MULHER"])]

    if nomeCSV == 'primeiro-semestre-2021.csv':
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[9].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas
        tabela = tabela.drop_duplicates(subset='Hash', keep='first')

        tabela = tabela[tabela['Grupo vulnerável'].isin(["01. OUTRAS VIOLÊNCIAS CONTRA A MULHER", "02. VIOLÊNCIA DOMÉSTICA E FAMILIAR CONTRA A MULHER"])]

    if nomeCSV == 'segundo-semestre-2021.csv':
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[10].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas
        tabela = tabela.drop_duplicates(subset='Hash', keep='first')

        tabela = tabela[tabela['Grupo vulnerável'].isin(["01. OUTRAS VIOLÊNCIAS CONTRA A MULHER", "02. VIOLÊNCIA DOMÉSTICA E FAMILIAR CONTRA A MULHER"])]

    if nomeCSV == 'primeiro-semestre-2022.csv':
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[11].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas
        tabela = tabela.drop_duplicates(subset='Hash', keep='first')

        tabela = tabela[tabela['Grupo vulnerável'].isin(["01. OUTRAS VIOLÊNCIAS CONTRA A MULHER", "02. VIOLÊNCIA DOMÉSTICA E FAMILIAR CONTRA A MULHER"])]

    if nomeCSV == 'segundo-semestre-2022.csv':
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[12].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas
        tabela = tabela.drop_duplicates(subset='Hash', keep='first')

        tabela = tabela[tabela['Grupo vulnerável'].isin(["OUTRAS VIOLÊNCIAS CONTRA A MULHER", "VIOLÊNCIA CONTRA A MULHER"])]

    if nomeCSV == 'primeiro-semestre-2023.csv':
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[13].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas
        tabela = tabela.drop_duplicates(subset='Hash', keep='first')

        tabela = tabela[tabela['Grupo vulnerável'].isin(["VIOLÊNCIA CONTRA A MULHER"])]

    if nomeCSV == 'segundo-semestre-2023.csv':
        novosNomesParaAsColunas = tabelaComNomesDasColunas.iloc[14].tolist()
        novosNomesParaAsColunas.pop(0)
        novosNomesParaAsColunas = [item for item in novosNomesParaAsColunas if not pd.isna(item)]
        tabela.columns = novosNomesParaAsColunas
        tabela = tabela.drop_duplicates(subset='Hash', keep='first')

    # Aplica sobre cada célula da tabela a fução que remove acentos e converte para caixa alta
    tabela = tabela.applymap(RemoveAcentosCaixaAlta)

    # Exporta para arquivo XLSX
    tabela.to_excel(nomeXLSX, index=False, engine='openpyxl')
```

```python
# Le o arquivo com o nome das colunas a serem subistituidas
tabelaComNomesDasColunas = pd.read_excel("Comparação dos campos de cada tabela.xlsx")

# Aplica a função para cada arquivo
Processa('ligue180-2014.csv', '../Etapa 2/2014.xlsx', tabelaComNomesDasColunas, ';')
print('2014 processado')

Processa('ligue180-2015.csv', '../Etapa 2/2015.xlsx', tabelaComNomesDasColunas, ';')
print('2015 processado')

Processa('ligue180-2016.csv', '../Etapa 2/2016.xlsx', tabelaComNomesDasColunas, ';')
print('2016 processado')

Processa('ligue180-2017.csv', '../Etapa 2/2017.xlsx', tabelaComNomesDasColunas, ';')
print('2017 processado')

Processa('ligue180-nov-2018.csv', '../Etapa 2/2018-1.xlsx', tabelaComNomesDasColunas, ';')
print('2018-1 processado')

Processa('ligue180-2018.csv', '../Etapa 2/2018-2.xlsx', tabelaComNomesDasColunas, ';')
print('2018-2 processado')

Processa('ligue180-2019.csv', '../Etapa 2/2019.xlsx', tabelaComNomesDasColunas, ';')
print('2019 processado')

Processa('primeiro-semestre-2020.csv', '../Etapa 2/2020-1.xlsx', tabelaComNomesDasColunas, ',')
print('2020-1 processado')

Processa('segundo-semestre-2020.csv', '../Etapa 2/2020-2.xlsx', tabelaComNomesDasColunas, ',')
print('2020-2 processado')

Processa('primeiro-semestre-2021.csv', '../Etapa 2/2021-1.xlsx', tabelaComNomesDasColunas, ',')
print('2021-1 processado')

Processa('segundo-semestre-2021.csv', '../Etapa 2/2021-2.xlsx', tabelaComNomesDasColunas, ',')
print('2021-2 processado')

Processa('primeiro-semestre-2022.csv', '../Etapa 2/2022-1.xlsx', tabelaComNomesDasColunas, ',')
print('2022-1 processado')

Processa('segundo-semestre-2022.csv', '../Etapa 2/2022-2.xlsx', tabelaComNomesDasColunas, ';')
print('2022-2 processado')

Processa('primeiro-semestre-2023.csv', '../Etapa 2/2023-1.xlsx', tabelaComNomesDasColunas, ';')
print('2023-1 processado')

Processa('segundo-semestre-2023.csv', '../Etapa 2/2023-2.xlsx', tabelaComNomesDasColunas, ';')
print('2023-2 processado')

```
